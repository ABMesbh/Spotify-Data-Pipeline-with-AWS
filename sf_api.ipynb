```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import os
```

```python
Credentiels = SpotifyClientCredentials(client_id=os.getenv('SPOTIPY_CLIENT_ID'), client_secret=os.getenv('SPOTIPY_CLIENT_SECRET'))
```

```python
sp = spotipy.Spotify(client_credentials_manager=Credentiels)
```

```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZF1DXcBWIGoYBM5M"
```

```python
api = playlist_link.split('/')[-1]
```

```python
data = sp.playlist_tracks(api)
```

```python
data
```

```python
data_dict = []
for record in data['items']:
    record_id = record['track']['id']
    record_artist = record['track']['artists'][0]['name']
    record_name = record['track']['name']
    record_popularity = record['track']['popularity']
    record_duration = record['track']['duration_ms']
    record_release_date = record['track']['album']['release_date']
    data_dict.append({'record_id' : record_id,'artist': record_artist, 'name': record_name, 'popularity': record_popularity, 'duration': record_duration, 'release_date': record_release_date})

```

```python
df = pd.DataFrame(data_dict)
```
